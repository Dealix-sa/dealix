```python
# OpenRouter Company OS Prototype
This notebook builds a sample company operating system platform with OpenRouter integration. It includes environment loading, validation, business module scaffolding, workflow orchestration, and scenario testing.
</VSCode.Cell>
<VSCode.Cell language="python">
# 1. Import Required Libraries
import os
import json
from dataclasses import dataclass, field
from typing import Any, Dict, List, Optional

import requests
</VSCode.Cell>
<VSCode.Cell language="python">
# 2. Load OpenRouter Environment Variables
OPENROUTER_API_KEY = os.getenv("OPENROUTER_API_KEY")
OPENROUTER_BASE_URL = os.getenv("OPENROUTER_BASE_URL", "https://openrouter.ai/api/v1")
OPENROUTER_MODEL = os.getenv("OPENROUTER_MODEL", "deepseek/deepseek-chat")
OPENROUTER_TIMEOUT = int(os.getenv("OPENROUTER_TIMEOUT", "120"))
PRIMARY_MODEL = os.getenv("PRIMARY_MODEL", "minimax/minimax-m2.5")
ARCHITECT_MODEL = os.getenv("ARCHITECT_MODEL", "minimax/minimax-m2.7")
LIGHT_MODEL = os.getenv("LIGHT_MODEL", "minimax/minimax-m2.1")
FALLBACK_MODEL = os.getenv("FALLBACK_MODEL", "deepseek/deepseek-chat")

assert OPENROUTER_API_KEY, "OPENROUTER_API_KEY is required"
print("OpenRouter env loaded:")
print("- base_url:", OPENROUTER_BASE_URL)
print("- model:", OPENROUTER_MODEL)
print("- timeout:", OPENROUTER_TIMEOUT)
</VSCode.Cell>
<VSCode.Cell language="python">
# 3. Validate OpenRouter API Access
health_url = f"{OPENROUTER_BASE_URL.rstrip('/')}/auth/key"
response = requests.get(
    health_url,
    headers={"Authorization": f"Bearer {OPENROUTER_API_KEY}"},
    timeout=10,
)
response.raise_for_status()
status = response.json()
print("OpenRouter connectivity OK")
print(json.dumps(status, indent=2))
</VSCode.Cell>
<VSCode.Cell language="python">
# 4. Define Business Operating System Modules
@dataclass
class CRMModule:
    customers: List[Dict[str, Any]] = field(default_factory=list)

    def add_customer(self, customer: Dict[str, Any]) -> None:
        self.customers.append(customer)

    def list_customers(self) -> List[Dict[str, Any]]:
        return self.customers


@dataclass
class HRModule:
    employees: List[Dict[str, Any]] = field(default_factory=list)

    def add_employee(self, employee: Dict[str, Any]) -> None:
        self.employees.append(employee)

    def list_employees(self) -> List[Dict[str, Any]]:
        return self.employees


@dataclass
class FinanceModule:
    budgets: Dict[str, float] = field(default_factory=dict)

    def set_budget(self, area: str, amount: float) -> None:
        self.budgets[area] = amount

    def get_budget(self, area: str) -> float:
        return self.budgets.get(area, 0.0)


@dataclass
class OperationsModule:
    projects: List[Dict[str, Any]] = field(default_factory=list)

    def add_project(self, project: Dict[str, Any]) -> None:
        self.projects.append(project)

    def list_projects(self) -> List[Dict[str, Any]]:
        return self.projects


@dataclass
class ReportingModule:
    reports: List[Dict[str, Any]] = field(default_factory=list)

    def add_report(self, report: Dict[str, Any]) -> None:
        self.reports.append(report)

    def latest_report(self) -> Optional[Dict[str, Any]]:
        return self.reports[-1] if self.reports else None
</VSCode.Cell>
<VSCode.Cell language="python">
# 5. Implement Core Company OS Workflows
@dataclass
class CompanyOS:
    crm: CRMModule = field(default_factory=CRMModule)
    hr: HRModule = field(default_factory=HRModule)
    finance: FinanceModule = field(default_factory=FinanceModule)
    operations: OperationsModule = field(default_factory=OperationsModule)
    reporting: ReportingModule = field(default_factory=ReportingModule)

    def onboard_customer(self, customer: Dict[str, Any]) -> None:
        self.crm.add_customer(customer)
        project = {
            "name": f"Onboard {customer['company']}",
            "customer": customer["company"],
            "status": "active",
        }
        self.operations.add_project(project)

    def hire_employee(self, employee: Dict[str, Any]) -> None:
        self.hr.add_employee(employee)
        self.finance.set_budget("headcount", self.finance.get_budget("headcount") + employee.get("salary", 0.0))

    def produce_report(self, title: str, body: str) -> None:
        self.reporting.add_report({"title": title, "body": body})

    def run_business_review(self) -> Dict[str, Any]:
        return {
            "customers": len(self.crm.list_customers()),
            "employees": len(self.hr.list_employees()),
            "projects": len(self.operations.list_projects()),
            "budget_headcount": self.finance.get_budget("headcount"),
        }


company = CompanyOS()
company.onboard_customer({"company": "Dealix Enterprise", "industry": "enterprise SaaS"})
company.hire_employee({"name": "Salman", "role": "CEO", "salary": 320000.0})
company.produce_report("Quarterly Strategy", "Build a Saudi-ready enterprise operating system.")
print(company.run_business_review())
</VSCode.Cell>
<VSCode.Cell language="python">
# 6. Integrate OpenRouter for Automation

def call_openrouter(prompt: str, model: str = OPENROUTER_MODEL) -> Dict[str, Any]:
    payload = {
        "model": model,
        "input": prompt,
        "temperature": 0.2,
        "max_output_tokens": 512,
    }
    response = requests.post(
        f"{OPENROUTER_BASE_URL.rstrip('/')}/responses",
        headers={
            "Authorization": f"Bearer {OPENROUTER_API_KEY}",
            "Content-Type": "application/json",
        },
        json=payload,
        timeout=OPENROUTER_TIMEOUT,
    )
    response.raise_for_status()
    return response.json()


prompt = (
    "Generate a concise executive summary for a Saudi enterprise SaaS platform that focuses on billing, onboarding, "
    "and company operating system readiness. Include a compliance and localization note."
)
openrouter_result = call_openrouter(prompt)
print(json.dumps(openrouter_result, indent=2)[:1000])
</VSCode.Cell>
<VSCode.Cell language="python">
# 7. Run Sample Business Scenario Tests
assert len(company.crm.list_customers()) == 1
assert len(company.hr.list_employees()) == 1
assert len(company.operations.list_projects()) == 1
assert company.reporting.latest_report()["title"] == "Quarterly Strategy"

print("Sample business scenario assertions passed.")
```

## 8. Add Compliance & Localization (Saudi Market)

```python
@dataclass
class ComplianceModule:
    """Saudi enterprise compliance: VAT, PDPL, contracts, audits."""
    vat_rate: float = 0.15
    company_cr: str = ""
    audit_logs: List[Dict[str, Any]] = field(default_factory=list)
    
    def validate_cr(self, cr_number: str) -> bool:
        """Validate Saudi CR number format."""
        if len(cr_number) != 10 or not cr_number.isdigit():
            return False
        first_digit = int(cr_number[0])
        return first_digit in (1, 2, 4, 6)  # Valid first digits for Saudi CR
    
    def calculate_vat(self, amount: float) -> Dict[str, float]:
        """Calculate VAT for invoice."""
        vat = amount * self.vat_rate
        return {"subtotal": amount, "vat": vat, "total": amount + vat}
    
    def log_audit(self, action: str, details: Dict[str, Any]) -> None:
        """Log compliance audit trail."""
        self.audit_logs.append({
            "timestamp": __import__("datetime").datetime.now().isoformat(),
            "action": action,
            "details": details,
        })


@dataclass
class LocalizationModule:
    """Support Arabic and English for Saudi market."""
    current_lang: str = "ar"  # ar or en
    translations: Dict[str, Dict[str, str]] = field(default_factory=dict)
    
    def __post_init__(self):
        self.translations = {
            "welcome": {"ar": "أهلا وسهلا", "en": "Welcome"},
            "invoice": {"ar": "فاتورة", "en": "Invoice"},
            "payment": {"ar": "الدفع", "en": "Payment"},
            "compliance": {"ar": "الامتثال", "en": "Compliance"},
            "company": {"ar": "الشركة", "en": "Company"},
        }
    
    def translate(self, key: str) -> str:
        """Get translated text."""
        return self.translations.get(key, {}).get(self.current_lang, key)
    
    def switch_language(self, lang: str) -> None:
        """Switch to Arabic or English."""
        if lang in ("ar", "en"):
            self.current_lang = lang


compliance = ComplianceModule()
compliance.company_cr = "1010012345"
assert compliance.validate_cr("1010012345") == True
print("✓ Saudi CR validation works")

localization = LocalizationModule()
print(f"Welcome (AR): {localization.translate('welcome')}")
localization.switch_language("en")
print(f"Welcome (EN): {localization.translate('welcome')}")
```

```python
@dataclass
class IntegratedCompanyOS:
    """Master orchestrator integrating all company OS modules."""
    crm: CRMModule
    hr: HRModule
    finance: FinanceModule
    ops: OperationsModule
    reporting: ReportingModule
    compliance: ComplianceModule
    localization: LocalizationModule
    analytics: AnalyticsModule
    
    def onboard_customer_full(self, company_name: str, sector: str) -> Dict[str, Any]:
        """Complete customer onboarding with compliance checks."""
        # CRM onboard
        cust_id = self.crm.add_customer({"name": company_name, "sector": sector})
        
        # Log compliance action
        self.compliance.log_audit("onboard_customer", {"company": company_name, "sector": sector})
        
        # Record in analytics
        self.analytics.add_kpi(f"customer_{cust_id}_onboard", 1.0)
        
        return {
            "customer_id": cust_id,
            "onboarded": True,
            "compliance_logged": True,
            "kpi_recorded": True,
            "message": self.localization.translate("welcome"),
        }
    
    def full_invoice_workflow(self, customer_id: str, amount: float) -> Dict[str, Any]:
        """Create invoice with VAT, compliance, and localization."""
        # Calculate VAT
        invoice_data = self.compliance.calculate_vat(amount)
        
        # Record in finance
        transaction_id = self.finance.record_transaction(customer_id, amount, "invoice")
        
        # Log audit
        self.compliance.log_audit("invoice_created", {
            "transaction_id": transaction_id,
            "amount": amount,
            "vat": invoice_data["vat"],
        })
        
        return {
            "transaction_id": transaction_id,
            "invoice_label": self.localization.translate("invoice"),
            "subtotal": invoice_data["subtotal"],
            "vat": invoice_data["vat"],
            "total": invoice_data["total"],
        }
    
    def generate_compliance_report(self) -> str:
        """Generate audit report in current language."""
        lang = self.localization.current_lang
        report_title = self.localization.translate("compliance")
        audit_count = len(self.compliance.audit_logs)
        return f"{report_title} Report ({lang}): {audit_count} audit events logged"
    
    def health_snapshot(self) -> Dict[str, Any]:
        """Return full company health snapshot."""
        return {
            "company_health": self.analytics.calculate_health_score(),
            "customers": len(self.crm.customers),
            "employees": len(self.hr.employees),
            "financial_balance": self.finance.get_balance(),
            "operations": self.ops.get_status(),
            "language": self.localization.current_lang,
            "compliance_events": len(self.compliance.audit_logs),
        }


# Instantiate integrated system
integrated_os = IntegratedCompanyOS(
    crm=CRMModule(),
    hr=HRModule(),
    finance=FinanceModule(),
    ops=OperationsModule(),
    reporting=ReportingModule(),
    compliance=ComplianceModule(),
    localization=LocalizationModule(),
    analytics=AnalyticsModule(),
)

# Onboard customer with compliance
result = integrated_os.onboard_customer_full("Dealix Enterprise", "Technology")
print("✓ Customer Onboard:", result)

# Create invoice
invoice = integrated_os.full_invoice_workflow("cust_001", 1000000.0)
print("✓ Invoice Created:", invoice)

# Generate report
report = integrated_os.generate_compliance_report()
print(f"✓ Report: {report}")

# Full health snapshot
health = integrated_os.health_snapshot()
print("✓ Company Health:", health)
```

## 11. AI-Powered Business Insights with OpenRouter

```python
def generate_ai_business_insights(company_os: IntegratedCompanyOS) -> str:
    """Generate AI-powered strategic insights using OpenRouter."""
    health = company_os.health_snapshot()
    api_key = os.getenv("OPENROUTER_API_KEY")
    if not api_key:
        return "⚠ OPENROUTER_API_KEY not set; skipping AI insights"
    
    prompt = f"""
Based on this company health snapshot, provide 3 strategic recommendations:
- Company Health Score: {health['company_health']:.1f}/100
- Customers: {health['customers']}
- Employees: {health['employees']}
- Financial Balance: {health['financial_balance']}
- Operations Status: {health['operations']}

For a Saudi enterprise operating in the technology sector, what are the top 3 priorities?
"""
    
    response = call_openrouter(
        model="minimax/minimax-m2.5",
        messages=[{"role": "user", "content": prompt}],
        max_tokens=500
    )
    
    return response


# Generate AI insights
print("Generating AI-powered insights from OpenRouter...")
insights = generate_ai_business_insights(integrated_os)
print("Strategic Insights:")
print(insights)
print()
```

```python
@dataclass
class AnalyticsModule:
    """Business intelligence and KPI dashboards."""
    kpis: Dict[str, float] = field(default_factory=dict)
    
    def add_kpi(self, name: str, value: float) -> None:
        """Register a KPI metric."""
        self.kpis[name] = value
    
    def get_dashboard_snapshot(self) -> Dict[str, Any]:
        """Return current dashboard state."""
        return {
            "kpi_count": len(self.kpis),
            "top_kpis": dict(sorted(self.kpis.items(), key=lambda x: x[1], reverse=True)[:5]),
            "generated_at": __import__("datetime").datetime.now().isoformat(),
        }
    
    def calculate_health_score(self) -> float:
        """Derive company health from KPIs (0–100)."""
        if not self.kpis:
            return 50.0
        normalized = [min(v / 100, 1.0) for v in self.kpis.values()]
        return sum(normalized) / len(normalized) * 100


analytics = AnalyticsModule()
analytics.add_kpi("revenue_growth", 120)
analytics.add_kpi("customer_retention", 95)
analytics.add_kpi("employee_satisfaction", 85)
analytics.add_kpi("operational_efficiency", 78)

dashboard = analytics.get_dashboard_snapshot()
health = analytics.calculate_health_score()
print(f"Dashboard: {dashboard}")
print(f"Company Health Score: {health:.1f}/100")
```
